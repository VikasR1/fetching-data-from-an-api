```python
import pandas as pd
import requests
```

```python
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/106.0.0.0 Safari/537.36',
  
}

params = {
    'genre_id': '2201',
    'widgets': 'ajax_widget_recommendations',
    'limit': '20',
    'output_type': 'json',
}

# calling api
response = requests.get('https://www.junodownload.com/api/1.2/widget/getProductListWidgets/', params=params, headers=headers)
```

```python
response.status_code
```

### convert the incoming data into json format

```python
get_json_response = response.json()
```

## Let's Convert the list in Dataframe

```python
temp_df = pd.DataFrame(get_json_response[0]['items'])
```

```python
temp_df.head(2)
```

we have everything which was available on API. Now, we need to select the items(columns) which we want

```python
Complete_df = temp_df[['title', 'title_full', 'artist', 'link', 'label', 'image_url']]
```

```python
# this is just one page data with 17 entries 
Complete_df
```

##### lets grab all pages data
for this we will create a loop and store that data in a `DF`, and then will append it with new `DF` .
Thus, once the loop will over, we will have a large `DataFrame` with complete data

```python
# create empty Df
df = pd.DataFrame()
```

```python
df
```

```python
for i in range(1, 201):
    response = requests.get('https://www.junodownload.com/api/{}'.format(i))
    temp_df = pd.DataFrame(get_json_response[0]['items'])
    com_df = temp_df[['title', 'title_full', 'artist', 'link', 'label', 'image_url']]
    df = df.append(com_df, ignore_index=True)
```

```python
# after completing the process verify the df size
df.shape
```

```python
# save the file in csv
df.to_csv('juno_data.csv')
```
